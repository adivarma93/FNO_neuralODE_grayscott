```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from tqdm.notebook import tqdm # Use notebook tqdm for better Jupyter integration
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torch.fft
import torch.nn.functional as F
import gc # Garbage collector

# --- Configuration ---
NX, NY = 128, 128 # Spatial resolution (For values of F and k, this Nx, Ny is required to develop patterns. 64x64 doesnt work here)
N_VARS = 2 # u, v

# Gray-Scott Parameters chosen from literature to get patterns
Du = 2e-5
Dv = 1e-5
# --- Use parameters known for complex patterns ---
F = 0.0545
k = 0.062

print(f"Using Parameters: Du={Du:.2e}, Dv={Dv:.2e}, F={F:.4f}, k={k:.4f}")
print(f"Grid Size: {NX}x{NY}")
print("--- Using HYBRID FNO Approach ---") 

# Domain & Time
L = 2.5  # Domain size [0, L] x [0, L]

T_END_GROUND_TRUTH = 10000 # Time for ground truth data generation (ensure patterns develop. it will eventually fill the grid)
T_END_SIM = 10000         # Time for FNO simulation (match or shorter)
N_FRAMES_GROUND_TRUTH = 1000 # Increased frames for better dynamics capture
N_FRAMES_SIM = 1000        # Number of frames for FNO simulation output

# --- FNO Configuration ---
FNO_MODES = 36  # adjust modes 
FNO_WIDTH = 64  #  width
N_EPOCHS = 150   # Training epochs 
BATCH_SIZE = 4 # Might need smaller batch size with higher res/width
LEARNING_RATE = 5e-5 # Adjusted learning rate
TRAIN_SPLIT = 0.8 # 

# --- Device Selection ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# ============================================================
# Part 1: Ground Truth Simulation (using original code)
# ============================================================
print("\n--- Part 1: Generating Ground Truth Data ---")

# --- Spectral RHS Function (Gray-Scott 2D) ---
def get_wavenumbers_2d(nx, ny, Lx, Ly):
    kx = np.fft.fftfreq(nx, d=Lx / (2 * np.pi * nx))
    ky = np.fft.fftfreq(ny, d=Ly / (2 * np.pi * ny))
    kx_grid, ky_grid = np.meshgrid(kx, ky, indexing='ij')
    k_sq = kx_grid**2 + ky_grid**2
    return k_sq

K_SQ_2D_NP = get_wavenumbers_2d(NX, NY, L, L)

def gray_scott_rhs_spectral_2d(t, state_flat, Du, Dv, F, k, k_sq_2d, nx, ny):
    # (Same as before)
    try:
        state = state_flat.reshape((N_VARS, nx, ny))
    except ValueError as e:
        print(f"Error reshaping state_flat: expected size {N_VARS*nx*ny}, got {state_flat.size}")
        raise e
    u, v = state[0, :, :], state[1, :, :]
    u_hat = np.fft.fft2(u)
    v_hat = np.fft.fft2(v)
    laplacian_u = np.fft.ifft2(u_hat * (-k_sq_2d)).real
    laplacian_v = np.fft.ifft2(v_hat * (-k_sq_2d)).real
    reaction_uv_sq = u * (v**2)
    dudt_react = -reaction_uv_sq + F * (1.0 - u)
    dvdt_react = reaction_uv_sq - (F + k) * v
    dudt = Du * laplacian_u + dudt_react
    dvdt = Dv * laplacian_v + dvdt_react
    rhs_flat = np.stack((dudt, dvdt), axis=0).flatten()
    return rhs_flat

# --- Initial Condition ---
# (Same as before)
print("Setting up initial condition...")
u0 = np.ones((NX, NY), dtype=np.float64) * 1.0
v0 = np.zeros((NX, NY), dtype=np.float64) * 0.0
pert_radius_outer = L / 8
cx, cy = L / 2, L / 2
x_ax = np.linspace(0, L, NX, endpoint=False)
y_ax = np.linspace(0, L, NY, endpoint=False)
X_grid, Y_grid = np.meshgrid(x_ax, y_ax, indexing='ij')
dist_sq = (X_grid - cx)**2 + (Y_grid - cy)**2
mask = (dist_sq < pert_radius_outer**2)
u0[mask] = 0.50 + (np.random.rand(*u0[mask].shape) - 0.5) * 0.1
v0[mask] = 0.25 + (np.random.rand(*v0[mask].shape) - 0.5) * 0.1
initial_state_flat = np.stack((u0, v0), axis=0).flatten()
initial_state_structured = initial_state_flat.reshape(N_VARS, NX, NY)


# --- Time Integration (Ground Truth) ---
t_span_gt = [0, T_END_GROUND_TRUTH]
t_eval_gt = np.linspace(t_span_gt[0], t_span_gt[1], N_FRAMES_GROUND_TRUTH)

print(f"Starting ground truth simulation (t={t_span_gt[0]} to t={t_span_gt[1]})...")
start_time_gt = time.time()

sol_gt = solve_ivp(gray_scott_rhs_spectral_2d, t_span_gt, initial_state_flat,
                   args=(Du, Dv, F, k, K_SQ_2D_NP, NX, NY),
                   t_eval=t_eval_gt,
                   rtol=1e-7, atol=1e-8, # Slightly tighter tolerances for better GT
                   method='RK45')

end_time_gt = time.time()

if not sol_gt.success:
    print(f"!!! Ground Truth Simulation FAILED. Status: {sol_gt.status}, Message: {sol_gt.message} !!!")
    exit()
else:
    print(f"Ground truth simulation finished successfully in {end_time_gt - start_time_gt:.2f} seconds.")
    trajectory_gt = sol_gt.y.T.reshape(-1, N_VARS, NX, NY)
    times_gt = sol_gt.t
    print(f"Ground Truth Trajectory shape: {trajectory_gt.shape}")

# ============================================================
# Part 2: FNO Model Definition (PyTorch)
# ============================================================
print("\n--- Part 2: Defining FNO Model ---")
# (FNO Model Definition is unchanged)
class SpectralConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, modes1, modes2):
        super(SpectralConv2d, self).__init__()
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.modes1 = modes1
        self.modes2 = modes2
        self.scale = (1 / (in_channels * out_channels))
        self.weights1 = nn.Parameter(self.scale * torch.rand(in_channels, out_channels, self.modes1, self.modes2, dtype=torch.cfloat))
        self.weights2 = nn.Parameter(self.scale * torch.rand(in_channels, out_channels, self.modes1, self.modes2, dtype=torch.cfloat))

    def compl_mul2d(self, input, weights):
        return torch.einsum("bixy,ioxy->boxy", input, weights)

    def forward(self, x):
        batchsize = x.shape[0]
        x_ft = torch.fft.rfft2(x)
        out_ft = torch.zeros(batchsize, self.out_channels, x.size(-2), x.size(-1)//2 + 1, dtype=torch.cfloat, device=x.device)
        out_ft[:, :, :self.modes1, :self.modes2] = \
            self.compl_mul2d(x_ft[:, :, :self.modes1, :self.modes2], self.weights1)
        out_ft[:, :, -self.modes1:, :self.modes2] = \
            self.compl_mul2d(x_ft[:, :, -self.modes1:, :self.modes2], self.weights2)
        x = torch.fft.irfft2(out_ft, s=(x.size(-2), x.size(-1)))
        return x

class FNO2d(nn.Module):
    def __init__(self, modes1, modes2, width):
        super(FNO2d, self).__init__()
        self.modes1 = modes1
        self.modes2 = modes2
        self.width = width
        self.padding = 8
        self.fc0 = nn.Linear(N_VARS, self.width)
        self.conv0 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.w0 = nn.Conv2d(self.width, self.width, 1)
        self.conv1 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.w1 = nn.Conv2d(self.width, self.width, 1)
        self.conv2 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.w2 = nn.Conv2d(self.width, self.width, 1)
        self.conv3 = SpectralConv2d(self.width, self.width, self.modes1, self.modes2)
        self.w3 = nn.Conv2d(self.width, self.width, 1)
        self.fc1 = nn.Linear(self.width, 128)
        self.fc2 = nn.Linear(128, N_VARS) 

    def forward(self, x):
        # x shape: (batch, channels, nx, ny)
        x = x.permute(0, 2, 3, 1)
        x = self.fc0(x)
        x = x.permute(0, 3, 1, 2)
        # x = F.pad(x, [0, self.padding, 0, self.padding]) # Optional padding

        x1 = self.conv0(x)
        x2 = self.w0(x)
        x = x1 + x2
        x = torch.nn.functional.gelu(x)

        x1 = self.conv1(x)
        x2 = self.w1(x)
        x = x1 + x2
        x = torch.nn.functional.gelu(x)

        x1 = self.conv2(x)
        x2 = self.w2(x)
        x = x1 + x2
        x = torch.nn.functional.gelu(x)

        x1 = self.conv3(x)
        x2 = self.w3(x)
        x = x1 + x2

        # x = x[..., :-self.padding, :-self.padding] # maybe Optional unpadding
        x = x.permute(0, 2, 3, 1)
        x = self.fc1(x)
        x = torch.nn.functional.gelu(x)
        x = self.fc2(x)
        x = x.permute(0, 3, 1, 2) # Back to (batch, channels, nx, ny)
        return x

# ============================================================
# Part 3: Prepare Training Data (Modified for Hybrid)
# ============================================================
print("\n--- Part 3: Preparing Training Data (Hybrid: Target = Diffusion Part) ---")

# Calculate the true FULL RHS (derivatives) for each state
rhs_gt = np.zeros_like(trajectory_gt)
print("Calculating true FULL RHS for training data...")
for i in tqdm(range(len(times_gt))):
    state_flat_i = trajectory_gt[i].flatten()
    rhs_flat_i = gray_scott_rhs_spectral_2d(times_gt[i], state_flat_i, Du, Dv, F, k, K_SQ_2D_NP, NX, NY)
    rhs_gt[i] = rhs_flat_i.reshape(N_VARS, NX, NY)

# Calculate the analytical reaction term for each state
print("Calculating analytical REACTION term for training data...")
reaction_gt = np.zeros_like(trajectory_gt)
for i in tqdm(range(len(times_gt))):
    u, v = trajectory_gt[i, 0], trajectory_gt[i, 1]
    reaction_uv_sq = u * (v**2)
    react_u = -reaction_uv_sq + F * (1.0 - u)
    react_v = reaction_uv_sq - (F + k) * v
    reaction_gt[i, 0] = react_u
    reaction_gt[i, 1] = react_v

# Define the target for the FNO: Diffusion = Full RHS - Reaction
target_diffusion_gt = rhs_gt - reaction_gt
print(f"Shape of target (diffusion part): {target_diffusion_gt.shape}")

# --- Standardize the target data (Diffusion Part) ---
target_diffusion_tensor = torch.tensor(target_diffusion_gt, dtype=torch.float32)
# Calculate mean/std per channel (axis 0=samples, 1=channel, 2=nx, 3=ny)
# Flatten spatial dims for calculation: (samples, channels, nx*ny)
flat_target = target_diffusion_tensor.view(target_diffusion_tensor.shape[0], N_VARS, -1)
Y_mean_diffusion = torch.mean(flat_target, dim=(0, 2)) # Shape [N_VARS]
Y_std_diffusion = torch.std(flat_target, dim=(0, 2))   # Shape [N_VARS]
# Add small epsilon to std dev to avoid division by zero
Y_std_diffusion = torch.max(Y_std_diffusion, torch.tensor(1e-8))

print(f"Target (Diffusion) Mean (U, V): {Y_mean_diffusion.tolist()}")
print(f"Target (Diffusion) StdDev (U, V): {Y_std_diffusion.tolist()}")

# Reshape mean/std to (1, N_VARS, 1, 1) for broadcasting
Y_data_standardized = (target_diffusion_tensor - Y_mean_diffusion.view(1, N_VARS, 1, 1)) / Y_std_diffusion.view(1, N_VARS, 1, 1)

# Input data remains the state trajectory
X_data = torch.tensor(trajectory_gt, dtype=torch.float32)

print(f"X_data shape (States): {X_data.shape}")
print(f"Y_data shape (Standardized Diffusion): {Y_data_standardized.shape}")

# Create a PyTorch Dataset
class GrayScottDataset(Dataset):
    def __init__(self, states, rhs_diffusion_standardized):
        self.states = states
        self.rhs_diffusion_standardized = rhs_diffusion_standardized

    def __len__(self):
        return self.states.shape[0]

    def __getitem__(self, idx):
        return self.states[idx], self.rhs_diffusion_standardized[idx]

# Split data
dataset_size = len(X_data)
train_size = int(TRAIN_SPLIT * dataset_size)
val_size = dataset_size - train_size
indices = torch.randperm(dataset_size).tolist() # Use torch shuffle
train_indices = indices[:train_size]
val_indices = indices[train_size:]

train_dataset = GrayScottDataset(X_data[train_indices], Y_data_standardized[train_indices])
val_dataset = GrayScottDataset(X_data[val_indices], Y_data_standardized[val_indices])

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print(f"Training samples: {len(train_dataset)}, Validation samples: {len(val_dataset)}")

# Clear large numpy arrays from memory
del trajectory_gt, rhs_gt, reaction_gt, target_diffusion_gt, target_diffusion_tensor
gc.collect()

# Move mean/std tensors needed for inference to the correct device and reshape
Y_mean_diffusion_inf = Y_mean_diffusion.to(device).view(1, N_VARS, 1, 1)
Y_std_diffusion_inf = Y_std_diffusion.to(device).view(1, N_VARS, 1, 1)


# ============================================================
# Part 4: Train the FNO Model
# ============================================================
print("\n--- Part 4: Training the FNO Model (Predicting Standardized Diffusion) ---")

fno_model = FNO2d(FNO_MODES, FNO_MODES, FNO_WIDTH).to(device)
optimizer = optim.Adam(fno_model.parameters(), lr=LEARNING_RATE, weight_decay=1e-5)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=N_EPOCHS)
criterion = nn.MSELoss() # Predict standardized diffusion part

train_losses = []
val_losses = []
best_val_loss = float('inf')

start_time_train = time.time()

for epoch in range(N_EPOCHS):
    fno_model.train()
    epoch_train_loss = 0.0
    pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{N_EPOCHS} [Train]', leave=False)
    for states, target_rhs_std in pbar:
        states = states.to(device, non_blocking=True)
        target_rhs_std = target_rhs_std.to(device, non_blocking=True)

        optimizer.zero_grad()
        pred_rhs_std = fno_model(states) # Predicts standardized diffusion
        loss = criterion(pred_rhs_std, target_rhs_std)
        loss.backward()
        optimizer.step()

        epoch_train_loss += loss.item() # Dataloader averages loss by default if drop_last=False

    # Average loss over batches
    epoch_train_loss /= len(train_loader)
    train_losses.append(epoch_train_loss)

    # Validation
    fno_model.eval()
    epoch_val_loss = 0.0
    with torch.no_grad():
        for states, target_rhs_std in val_loader:
            states = states.to(device, non_blocking=True)
            target_rhs_std = target_rhs_std.to(device, non_blocking=True)
            pred_rhs_std = fno_model(states)
            loss = criterion(pred_rhs_std, target_rhs_std)
            epoch_val_loss += loss.item()

    epoch_val_loss /= len(val_loader)
    val_losses.append(epoch_val_loss)

    scheduler.step()

    print(f"Epoch {epoch+1}/{N_EPOCHS} - Train Loss: {epoch_train_loss:.4e}, Val Loss: {epoch_val_loss:.4e}, LR: {scheduler.get_last_lr()[0]:.2e}")

    # Optional: Save best model based on validation loss
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        torch.save(fno_model.state_dict(), 'best_fno_hybrid_model.pth')
        # print(f"  -> Saved new best model (Val Loss: {best_val_loss:.4e})")


end_time_train = time.time()
print(f"Training finished in {end_time_train - start_time_train:.2f} seconds.")
print(f"Best Validation Loss: {best_val_loss:.4e}")

# Optional: Load the best model for inference
# print("Loading best model for inference...")
# fno_model.load_state_dict(torch.load('best_fno_hybrid_model.pth'))
# fno_model.eval()

# Plot training history
plt.figure(figsize=(8, 4))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss (Standardized Diffusion)')
plt.yscale('log')
plt.title('FNO Hybrid Training History')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================
# Part 5: Neural ODE Simulation using Trained FNO (Hybrid)
# ============================================================
print("\n--- Part 5: Simulating with Hybrid FNO as Neural ODE ---")

fno_model.eval() # Ensure model is in evaluation mode

# Wrapper function for solve_ivp using the FNO for diffusion + analytical reaction
def fno_hybrid_rhs_wrapper(t, state_flat, model, device, nx, ny, F_param, k_param, y_mean_diff, y_std_diff):
    """
    Computes RHS using FNO for diffusion and analytical reaction.
    y_mean_diff, y_std_diff should be tensors on the correct device,
    shaped [1, N_VARS, 1, 1] for broadcasting.
    """
    # Reshape flat numpy array to structured torch tensor
    state_np = state_flat.reshape((N_VARS, nx, ny))
    state_tensor = torch.tensor(state_np, dtype=torch.float32).unsqueeze(0).to(device) # Add batch dim

    # --- FNO Prediction (Standardized Diffusion) ---
    with torch.no_grad():
        diffusion_pred_standardized = model(state_tensor)

    # --- Un-standardize the prediction ---
    diffusion_pred_tensor = diffusion_pred_standardized * y_std_diff + y_mean_diff
    diffusion_pred_np = diffusion_pred_tensor.squeeze(0).cpu().numpy() # Remove batch dim

    # --- Analytical Reaction Term ---
    u, v = state_np[0], state_np[1]
    reaction_uv_sq = u * (v**2)
    react_u = -reaction_uv_sq + F_param * (1.0 - u)
    react_v = reaction_uv_sq - (F_param + k_param) * v
    reaction_np = np.stack((react_u, react_v), axis=0) # Shape [N_VARS, nx, ny]

    # --- Combine FNO Diffusion + Analytical Reaction ---
    rhs_full_np = diffusion_pred_np + reaction_np

    # --- (Optional) Diagnostic ---
    # if t < 1.0 or np.random.rand() < 0.01:
    #      diff_norm = np.linalg.norm(diffusion_pred_np)
    #      react_norm = np.linalg.norm(reaction_np)
    #      total_norm = np.linalg.norm(rhs_full_np)
    #      print(f"t={t:.2f}, Norms: Diff={diff_norm:.3e}, React={react_norm:.3e}, Total={total_norm:.3e}")
    # --- End Diagnostic ---

    rhs_full_flat = rhs_full_np.flatten()
    return rhs_full_flat

# --- Time Integration (FNO Hybrid Neural ODE) ---
t_span_sim = [0, T_END_SIM]
# Ensure simulation times match frames for comparison plotting later
t_eval_sim = np.linspace(t_span_sim[0], t_span_sim[1], N_FRAMES_SIM)

print(f"Starting FNO Hybrid simulation (t={t_span_sim[0]} to t={t_span_sim[1]})...")
start_time_fno_sim = time.time()

initial_state_flat_fno = initial_state_flat.copy()

# Pass F, k, and the mean/std tensors for diffusion target to the solver args
sol_fno = solve_ivp(fno_hybrid_rhs_wrapper, t_span_sim, initial_state_flat_fno,
                    args=(fno_model, device, NX, NY, F, k, Y_mean_diffusion_inf, Y_std_diffusion_inf),
                    t_eval=t_eval_sim,
                    rtol=1e-5, atol=1e-6, # Can adjust tolerances
                    method='RK45') # Or try 'BDF' if it seems stiff

end_time_fno_sim = time.time()

if sol_fno.success:
    print(f"FNO Hybrid simulation finished successfully in {end_time_fno_sim - start_time_fno_sim:.2f} seconds.")
    trajectory_fno = sol_fno.y.T.reshape(-1, N_VARS, NX, NY)
    times_fno = sol_fno.t
    print(f"FNO Hybrid Trajectory shape: {trajectory_fno.shape}")
else:
    print(f"!!! FNO Hybrid Simulation FAILED. Status: {sol_fno.status}, Message: {sol_fno.message} !!!")
    trajectory_fno = None
    times_fno = None

# ============================================================
# Part 6: Comparison and Visualization
# ============================================================
print("\n--- Part 6: Comparing Ground Truth and FNO Hybrid Simulation ---")

if trajectory_fno is not None:
    # Regenerate ground truth for comparison at the FNO simulation time points
    print("Re-running ground truth simulation at FNO evaluation times for comparison...")
    sol_gt_comp = solve_ivp(gray_scott_rhs_spectral_2d, t_span_gt, initial_state_flat,
                           args=(Du, Dv, F, k, K_SQ_2D_NP, NX, NY),
                           t_eval=t_eval_sim, # Use FNO sim eval times
                           rtol=1e-7, atol=1e-8, method='RK45')

    if sol_gt_comp.success:
        trajectory_gt_comp = sol_gt_comp.y.T.reshape(-1, N_VARS, NX, NY)
        times_gt_comp = sol_gt_comp.t
        print(f"Comparison Ground Truth Trajectory shape: {trajectory_gt_comp.shape}")
        # Sanity check times
        if not np.allclose(times_fno, times_gt_comp):
             print("WARNING: Regenerated ground truth times do not exactly match FNO times!")
    else:
        print("Failed to regenerate ground truth for comparison.")
        trajectory_gt_comp = None

    # Proceed with plotting only if comparison ground truth was generated successfully
    if trajectory_gt_comp is not None:
        n_plots = min(N_FRAMES_SIM, 8)
        plot_indices = np.linspace(0, len(times_fno) - 1, n_plots, dtype=int)

        vmin_u = min(np.min(trajectory_gt_comp[plot_indices, 0]), np.min(trajectory_fno[plot_indices, 0]))
        vmax_u = max(np.max(trajectory_gt_comp[plot_indices, 0]), np.max(trajectory_fno[plot_indices, 0]))
        vmin_v = min(np.min(trajectory_gt_comp[plot_indices, 1]), np.min(trajectory_fno[plot_indices, 1]))
        vmax_v = max(np.max(trajectory_gt_comp[plot_indices, 1]), np.max(trajectory_fno[plot_indices, 1]))
        vmin_u = min(vmin_u, vmax_u - 1e-6)
        vmin_v = min(vmin_v, vmax_v - 1e-6)

        fig, axes = plt.subplots(N_VARS * 2, n_plots, figsize=(n_plots * 2.5, 9), sharex=True, sharey=True)
        fig.suptitle(f'Gray-Scott Comparison (F={F}, k={k}) - Top: Ground Truth, Bottom: FNO Hybrid', fontsize=14)

        for i, plot_idx in enumerate(plot_indices):
            t_plot = times_fno[plot_idx]

            # Plot Ground Truth U/V
            im_u_gt = axes[0, i].imshow(trajectory_gt_comp[plot_idx, 0, :, :], origin='lower', cmap='viridis', vmin=vmin_u, vmax=vmax_u, extent=[0, L, 0, L])
            axes[0, i].set_title(f'GT U (t={t_plot:.0f})')
            if i == 0: axes[0, i].set_ylabel('GT U')
            im_v_gt = axes[1, i].imshow(trajectory_gt_comp[plot_idx, 1, :, :], origin='lower', cmap='coolwarm', vmin=vmin_v, vmax=vmax_v, extent=[0, L, 0, L])
            axes[1, i].set_title(f'GT V (t={t_plot:.0f})')
            if i == 0: axes[1, i].set_ylabel('GT V')

            # Plot FNO Hybrid U/V
            im_u_fno = axes[2, i].imshow(trajectory_fno[plot_idx, 0, :, :], origin='lower', cmap='viridis', vmin=vmin_u, vmax=vmax_u, extent=[0, L, 0, L])
            axes[2, i].set_title(f'FNO U (t={t_plot:.0f})')
            if i == 0: axes[2, i].set_ylabel('FNO U')
            im_v_fno = axes[3, i].imshow(trajectory_fno[plot_idx, 1, :, :], origin='lower', cmap='coolwarm', vmin=vmin_v, vmax=vmax_v, extent=[0, L, 0, L])
            axes[3, i].set_title(f'FNO V (t={t_plot:.0f})')
            if i == 0: axes[3, i].set_ylabel('FNO V')

            for row_ax in axes[:, i]:
                row_ax.set_xticks([])
                row_ax.set_yticks([])

        #fig.colorbar(im_u_gt, ax=axes[[0, 2], :].ravel().tolist(), shrink=0.6, label='U Conc.', pad=0.02)
        #fig.colorbar(im_v_gt, ax=axes[[1, 3], :].ravel().tolist(), shrink=0.6, label='V Conc.', pad=0.02)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

        # Optional: Calculate and plot relative error over time
        if trajectory_gt_comp.shape == trajectory_fno.shape:
            norm_gt_u = np.linalg.norm(trajectory_gt_comp[:, 0], axis=(1, 2))
            norm_gt_v = np.linalg.norm(trajectory_gt_comp[:, 1], axis=(1, 2))
            valid_u = norm_gt_u > 1e-10
            valid_v = norm_gt_v > 1e-10
            error_u = np.full_like(norm_gt_u, np.nan)
            error_v = np.full_like(norm_gt_v, np.nan)
            error_u[valid_u] = np.linalg.norm(trajectory_gt_comp[valid_u, 0] - trajectory_fno[valid_u, 0], axis=(1, 2)) / norm_gt_u[valid_u]
            error_v[valid_v] = np.linalg.norm(trajectory_gt_comp[valid_v, 1] - trajectory_fno[valid_v, 1], axis=(1, 2)) / norm_gt_v[valid_v]

            plt.figure(figsize=(8, 4))
            plt.plot(times_fno[valid_u], error_u[valid_u], label='Relative Error U', marker='.')
            plt.plot(times_fno[valid_v], error_v[valid_v], label='Relative Error V', marker='.')
            plt.xlabel('Time')
            plt.ylabel('Relative L2 Error')
            plt.title('FNO Hybrid Simulation Error vs Ground Truth')
            plt.legend()
            plt.grid(True)
            plt.yscale('log')
            plt.tight_layout()
            plt.show()
        else:
             print("Error calculation skipped: Trajectory shapes mismatch.")

    else:
        print("Cannot create comparison plot as ground truth regeneration failed.")

else:
    print("Cannot visualize comparison as FNO simulation failed.")

print("\nFNO Hybrid Neural ODE integration script finished.")
```
